## Implemented Ideas scraping logic

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

list_of_post_links = []
loop_count = 1

while loop_count < 8: # manually entered this limit after checking website
    current_link = 'https://community.spotify.com/t5/forums/searchpage/tab/message?filter=includeIdeas,location&include_ideas=true&advanced=true&location=idea-board:ideas_implemented&page={0}&sort_by=-topicPostDate&collapse_discussion=true&search_type=thread&search_page_size=50'.format(loop_count)
    print("Entering Loop " + str(loop_count) + ":")
    driver = webdriver.Chrome(options=options)
    driver.get(current_link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    links_to_ideas = soup.find_all('a', class_="page-link lia-link-navigation lia-custom-event")
    for link in links_to_ideas:
        final_link = 'https://community.spotify.com' + link['href']
        print(final_link)
        list_of_post_links.append(final_link)
    soup_next = BeautifulSoup(driver.page_source, 'html.parser')

    driver.quit()
    time.sleep(5)
    loop_count = loop_count + 1

f = open("implemented_post_links.txt", "w")
for link in list_of_post_links:
    f.write(link + "\n")
f.close()
```

## Live Ideas scraping logic

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

list_of_post_links = []
loop_count = 1

while loop_count < 23: # manually entered this limit after checking website
    current_link = 'https://community.spotify.com/t5/Live-Ideas/idb-p/ideas_live/tab/most-recent/page/{0}'.format(loop_count)
    print("Entering Loop " + str(loop_count) + ":")
    driver = webdriver.Chrome(options=options)
    driver.get(current_link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    links_to_ideas = soup.find_all('a')

    for link in links_to_ideas:
        if link['href'].startswith("/t5/Live-Ideas") and link['href'] != '/t5/Live-Ideas/idb-p/ideas_live':
            final_link = 'https://community.spotify.com' + link['href']
            print(final_link)
            list_of_post_links.append(final_link)
    soup_next = BeautifulSoup(driver.page_source, 'html.parser')

    driver.quit()
    time.sleep(5)
    loop_count = loop_count + 1

f = open("live_idea_links.txt", "w")
for link in list_of_post_links:
    f.write(link + "\n")
f.close()
```

## Closed Ideas scraping logic

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

list_of_post_links = []
loop_count = 1

while loop_count <= 16: # manually entered this limit after checking website
    current_link = 'https://community.spotify.com/t5/Closed-Ideas/idb-p/ideas_no/tab/most-recent/page/{0}'.format(loop_count)
    print("Entering Loop " + str(loop_count) + ":")
    driver = webdriver.Chrome(options=options)
    driver.get(current_link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    links_to_ideas = soup.find_all('a')

    for link in links_to_ideas:
        try:
            if link['href'].startswith("/t5/Closed-Ideas") and link['href'] != '/t5/Closed-Ideas/idb-p/ideas_no':
                final_link = 'https://community.spotify.com' + link['href']
                print(final_link)
                list_of_post_links.append(final_link)
        except:
            continue
    soup_next = BeautifulSoup(driver.page_source, 'html.parser')

    driver.quit()
    time.sleep(5)
    loop_count = loop_count + 1

f = open("closed_idea_links.txt", "w")
for link in list_of_post_links:
    f.write(link + "\n")
f.close()
```
